```python
from PIL import Image

# ASCII characters based on brightness
ASCII_CHARS = ['@', '#', 'S', '%', '?', '*', '+', ';', ':', ',', '.']

# Resize the image
def resize_image(image, new_width=100):
    width, height = image.size
    aspect_ratio = height / width
    new_height = int(aspect_ratio * new_width)
    return image.resize((new_width, new_height))

# Convert the image to grayscale
def grayscale_image(image):
    return image.convert('L')

# Map pixels to ASCII characters
def pixels_to_ascii(image):
    pixels = image.getdata()
    ascii_str = ''.join([ASCII_CHARS[pixel // 25] for pixel in pixels])
    return ascii_str

# Main function to generate ASCII art
def generate_ascii(image_path):
    # Open the image
    try:
        image = Image.open(image_path)
    except Exception as e:
        print(f"Unable to open image file: {image_path}. {e}")
        return
    
    # Resize and convert to grayscale
    image = resize_image(image)
    image = grayscale_image(image)

    # Convert pixels to ASCII
    ascii_str = pixels_to_ascii(image)

    # Format ASCII string with proper line breaks
    img_width = image.width
    ascii_art = "\n".join([ascii_str[i:(i + img_width)] for i in range(0, len(ascii_str), img_width)])

    # Print or save ASCII art
    with open("ascii_image.txt", "w") as f:
        f.write(ascii_art)

    print(ascii_art)

# Example usage
generate_ascii("/Users/needhikore/Downloads/img.jpg")
```
